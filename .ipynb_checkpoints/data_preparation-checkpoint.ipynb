#  Data preparation

```python
COL_TYPES = [0]+ [''] * 2 + [0] * 4 
COL_NAMES = ["DocID","RawText","target"]
```

```python
AGE_CUT_VALUES = np.concatenate([[-2, -1, 1], (5 + np.arange(0, 125, 5))])
```

```python
df = pd.read_csv("data/Corpus.csv",sep=";",encoding='latin-1',low_memory=False,).rename(columns={"YearCoded":"year",
                                                                                  "Gender":"sex",
                                                                                  "Age":"age", 
                                                                                 "OriginalText" : "RawText",
                                                                                "RevisedText":"target",
                                                                                "DocId":"DocID"})

COL_DROP = ["LocationOfDeath","TypeIntervalLineCause","ValueInterval","RankLineCause","LineCause","ICDCode","DictionaryText"]

df = df.drop(COL_DROP, axis=1)




df
```

Rajouter le numéro de lignes (sous forme de token donc passer par une petite transformation) avant la concaténation à chaque phrases target afin de pouvoir bien situer les mots

```python
texts = df.groupby(["DocID","LineID"])["RawText"].unique().apply(lambda x: x[0])
sdt_texts = df.groupby(["DocID","LineID"])["target"].apply(lambda x: ' [SEPCODE] '.join(x))

texts
```

```python
sdt_texts
```

```python
prep_cepidc = pd.merge(texts,sdt_texts,left_index = True,right_index=True)
prep_cepidc = pd.merge(prep_cepidc,
                       df[["DocID","sex","year","age"]].groupby("DocID").first(),
                       left_index = True,right_index=True)

prep_cepidc = prep_cepidc.drop(["sex","year","age"], axis=1)
prep_cepidc
```

On enrengistre l'input et l'output dans deux fichiers csv en prenant bien soin à changer l'encoding afin de ne pas avoir de problème de caractère

```python
prep_cepidc['RawText'].to_csv("data/extrait500_corpus_TexteBrut.csv",index = False, header = False, encoding='latin-1')
prep_cepidc['target'].to_csv("data/extrait500_corpus_TextePropre.csv",index = False, header = False, encoding='latin-1')
```

```python
def get_dataset_partitions_pd(df, train_split=0.8, val_split=0.1, test_split=0.1):
    assert (train_split + test_split + val_split) == 1
    
    # Only allows for equal validation and test splits
    assert val_split == test_split 

    # Specify seed to always have the same split distribution between runs
    df_sample = df.sample(frac=1, random_state=12)
    train_i = int(train_split * len(df))
    val_i = int(val_split * len(df)) + train_i
    indices_or_sections = [train_i,val_i]
    
    train_ds, val_ds, test_ds = np.split(df_sample, indices_or_sections)
    
    return train_ds, val_ds, test_ds
```

```python
prep_cepidc["target"] = prep_cepidc.target
```

```python
sample_train,sample_val,sample_test = get_dataset_partitions_pd(prep_cepidc.loc[prep_cepidc.target.isin(selected_association_codes[:25]),COL_NAMES[1:]])


```

```python
import unidecode


# opening the file in read mode
file = open('data/extrait500_corpus_train_code_target.csv', "r")
replacement = ""
# using the for loop
for line in file:
    line1 = line.replace( '"', '')
    line2 = line1.rstrip( '\"')
    replacement = replacement + unidecode.unidecode(line2)

file.close()
# opening the file in write mode
fout = open('data/extrait500_corpus_train_code_target.csv', "w")
fout.write(replacement)
fout.close()
```
