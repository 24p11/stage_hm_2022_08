```python
df = pd.read_csv("data/Corpus.csv",sep=";",encoding='latin-1',low_memory=False).rename(columns={"YearCoded":"year",
                                                                                  "Gender":"sex",
                                                                                  "Age":"age", 
                                                                                 "OriginalText" : "RawText",
                                                                                "RevisedText":"target",
                                                                                "DocId":"DocID"})

COL_DROP = ["LocationOfDeath","TypeIntervalLineCause","ValueInterval","RankLineCause","LineCause","ICDCode","DictionaryText"]

df = df.drop(COL_DROP, axis=1)

df['RawText'] = df['RawText'].apply(unidecode.unidecode)
df['target'] = df['target'].apply(unidecode.unidecode)

df["RawText"]= df["RawText"].str.replace("\"","")
df["target"]=df["target"].str.replace("\"","")         

df
```

```python
df["ADAPT"] = df["RawText"] +" " + df["target"]
df
```

Rajouter le numéro de lignes (sous forme de token donc passer par une petite transformation) avant la concaténation à chaque phrases target afin de pouvoir bien situer les mots

```python
texts = df.groupby(["DocID","LineID"])["RawText"].unique().apply(lambda x: x[0])
sdt_texts = df.groupby(["DocID","LineID"])["target"].apply(lambda x: ' [SEPCODE] '.join(x))

texts
```

```python
prep_cepidc = pd.merge(texts,sdt_texts,left_index = True,right_index=True)
prep_cepidc = pd.merge(prep_cepidc,
                       df[["DocID","sex","year","age"]].groupby("DocID").first(),
                       left_index = True,right_index=True)

prep_cepidc = prep_cepidc.drop(["sex","year","age"], axis=1)
prep_cepidc
```

```python
def get_dataset_partitions_pd(df, train_split=0.8, val_split=0.1, test_split=0.1):
    assert (train_split + test_split + val_split) == 1
    
    # Only allows for equal validation and test splits
    assert val_split == test_split 

    # Specify seed to always have the same split distribution between runs
    df_sample = df.sample(frac=1, random_state=12)
    train_i = int(train_split * len(df))
    val_i = int(val_split * len(df)) + train_i
    indices_or_sections = [train_i,val_i]
    
    train_ds, val_ds, test_ds = np.split(df_sample, indices_or_sections)
    
    return train_ds, val_ds, test_ds
```

```python
sample_train,sample_val,sample_test = get_dataset_partitions_pd(prep_cepidc)


```

```python
sample_train.to_csv("data/Full_corpus_train_code_target.csv",sep=";", encoding='utf-8')
sample_val.to_csv("data/Full_corpus_val_code_target.csv",sep=";", encoding='utf-8')
sample_test.to_csv("data/Full_corpus_test_code_target.csv",sep=";", encoding='utf-8')
```
